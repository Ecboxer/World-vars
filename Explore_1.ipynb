Bring in country_codes.csv to add the region or sub-region feature

```python
codes = pd.read_csv('Country_codes.csv', delimiter = ',', encoding = 'utf-8-sig')
codes.head()
```

```python
country_codes = codes[['name', 'sub-region']]
country_codes.head()
```

```python
df_countries = df[['Country']]
codes_countries = codes[['name']]
df_countries.head()
```

```python
df_countries_array = df_countries.values.flatten()
codes_countries_array = codes_countries.values.flatten()
```

```python
common_countries = set(df_countries_array) & set(codes_countries_array)
missing_countries_l = set(df_countries_array) - set(codes_countries_array)
missing_countries_r = set(codes_countries_array) - set(df_countries_array)
len(missing_countries_l)
```

```python
def Shorten_strings(x):
    shorter = []
    for name in x:
        i = name.find(' ')
        if i == -1:
            shorter.append(name)
        else:    
            shorter.append(name[: i])
    return shorter
missing_l_abbr = Shorten_strings(missing_countries_l)
missing_r_abbr = Shorten_strings(missing_countries_r)
len(missing_l_abbr)
```

```python
common_abbr = set(missing_l_abbr) & set(missing_r_abbr)
missing_l_abbr_p = set(missing_l_abbr) - set(missing_r_abbr)
missing_r_abbr_p = set(missing_r_abbr) - set(missing_l_abbr)
len(missing_l_abbr_p)
```

Try to change the entries of missing_r_abbr_p to remove the diacritical marks

```python
import unidecode
missing_r_uni = []
for name in missing_r_abbr_p:
    name_uni = unidecode.unidecode(name)
    missing_r_uni.append(name_uni)
# missing_r_uni
```

```python
common_uni = set(missing_l_abbr_p) & set(missing_r_uni)
manual_l = set(missing_l_abbr_p) - set(missing_r_uni)
manual_r = set(missing_r_uni) - set(missing_l_abbr_p)
len(manual_l)
```
